```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import random

# Basis URL en Headers
# We gebruiken de URL voor seizoen 2024 (startjaar van seizoen 24/25)
URL = "https://www.transfermarkt.be/jupiler-pro-league/spieltag/wettbewerb/BE1/saison/2025/spieltag/"

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

def scrape_matchday(speeldag):
    print(f"--- Bezig met scrapen van speeldag {speeldag} ---")
    full_url = URL + str(speeldag)
    
    try:
        r = requests.get(full_url, headers=HEADERS)
        r.raise_for_status()
    except requests.exceptions.HTTPError as err:
        print(f"Fout bij ophalen speeldag {speeldag}: {err}")
        return []

    soup = BeautifulSoup(r.content, "html.parser")
    matchday_data = []

    # 1. Rode kaarten ophalen (Overzichtspagina)
    home_reds = []
    away_reds = []

    match_boxes = soup.find_all("div", class_="box")
    valid_boxes = [b for b in match_boxes if b.find("tr", class_="table-grosse-schrift")]

    for box in valid_boxes:
        home_rc = 0
        away_rc = 0
        
        spelverloop_row = box.find("tr", class_="spieltagsansicht-zwischenheadline")
        if spelverloop_row:
            current_row = spelverloop_row.find_next_sibling("tr")
            while current_row and "aktionsendzeile" not in current_row.get("class", []):
                home_cell = current_row.find("td", class_="rechts no-border-rechts spieltagsansicht")
                away_cell = current_row.find("td", class_="links no-border-links spieltagsansicht")

                if home_cell:
                    home_rc += len(home_cell.find_all("span", class_="icons_sprite icon-rotekarte-formation"))
                    home_rc += len(home_cell.find_all("span", class_="icons_sprite icon-gelbrotekarte-formation"))
                if away_cell:
                    away_rc += len(away_cell.find_all("span", class_="icons_sprite icon-rotekarte-formation"))
                    away_rc += len(away_cell.find_all("span", class_="icons_sprite icon-gelbrotekarte-formation"))

                current_row = current_row.find_next_sibling("tr")
        
        home_reds.append(home_rc)
        away_reds.append(away_rc)

    # 2. Teamnamen
    homes = []
    aways = []
    elements = soup.find_all("td", class_="spieltagsansicht-vereinsname")
    for i in range(0, len(elements), 4):
        if i+2 < len(elements):
            homes.append(elements[i].find("a").get_text(strip=True))
            aways.append(elements[i+2].find("a").get_text(strip=True))

    # 3. Rankings
    homesRankings = []
    awaysRankings = []
    rows = soup.find_all("tr", class_="table-grosse-schrift")
    for row in rows:
        rankings = row.find_all("span", class_="tabellenplatz")
        if len(rankings) >= 2:
            home_raw = rankings[0].get_text(strip=True)
            away_raw = rankings[-1].get_text(strip=True)
            homesRankings.append(home_raw.replace("(", "").replace(".)", "").replace(".", ""))
            awaysRankings.append(away_raw.replace("(", "").replace(".)", "").replace(".", ""))
        else:
            homesRankings.append("")
            awaysRankings.append("")

    # 4. Toeschouwers
    attendees = []
    icons = soup.find_all("span", class_="icon-zuschauer-zahl")
    for icon in icons:
        td = icon.parent
        attendees.append(td.get_text(" ", strip=True))

    # 5. Scheidsrechters
    referees = []
    ref_spans = [s for s in soup.find_all("span") if s.text and "Scheidsrechter" in s.text]
    for span in ref_spans:
        a_tag = span.find("a")
        referees.append(a_tag.get_text(strip=True) if a_tag else "Onbekend")

    # 6. Datum en Tijd
    date_cells = soup.find_all("td", class_="zentriert no-border", colspan="5")
    weekdays = []
    times = []
    for cell in date_cells:
        text = cell.get_text(" ", strip=True)
        if "uur" in text:
            left, right = text.split("-", 1)
            weekdays.append(left.strip().split()[0].replace(",",""))
            times.append(right.strip())

    # 7. Scores
    scores_home = []
    scores_away = []
    score_spans = soup.find_all("span", class_="matchresult finished")
    for span in score_spans:
        text = span.get_text(strip=True)
        if ":" in text:
            h, a = text.split(":", 1)
            scores_home.append(int(h))
            scores_away.append(int(a))
        else:
            scores_home.append(0)
            scores_away.append(0)

    # 8. Gedetailleerde statistieken (Deep dive per match)
    doelpogingen_home = []
    doelpogingen_away = []
    schoten_naast_home = []
    schoten_naast_away = []
    hoekschoppen_home = [] 
    hoekschoppen_away = [] 
    overtredingen_home = [] 
    overtredingen_away = [] 
    totale_marktwaarde_home = []
    totale_marktwaarde_away = []
    opstelling_home = []
    opstelling_away = []
    bank_home = []
    bank_away = []

    match_report_links = soup.find_all("a", title="Wedstrijdverslag", string="Wedstrijdverslag")
    
    for i, a_tag in enumerate(match_report_links):
        print(f"  > Ophalen details match {i+1}/{len(match_report_links)}...")
        href = a_tag.get("href")
        full_url_report = "https://www.transfermarkt.be" + href
        
        # Random delay tegen IP block
        time.sleep(random.uniform(1.0, 2.0))

        try:
            resp = requests.get(full_url_report, headers=HEADERS)
            soup_report = BeautifulSoup(resp.text, "html.parser")
            
            canonical_tag = soup_report.find("link", rel="canonical")
            if not canonical_tag:
                # Als de pagina niet goed laadt, voeg lege waarden toe en ga door
                raise Exception("Geen canonical link gevonden")

            canonical_url = canonical_tag['href']
            statistiek_url = canonical_url.replace("/index/", "/statistik/")
            vergelijken_url = statistiek_url.replace("statistik", "vorbericht")
            opstelling_url = statistiek_url.replace("statistik", "aufstellung")

            # --- STATISTIEKEN ---
            resp_stats = requests.get(statistiek_url, headers=HEADERS)
            soup_stats = BeautifulSoup(resp_stats.text, "html.parser")
            
            def get_stat(soup_obj, text_keyword):
                # Zoek naar de header die het keyword bevat
                section = soup_obj.find("div", class_="unterueberschrift", string=lambda t: t and text_keyword in t)
                if section:
                    div = section.find_next_sibling("div", class_="sb-statistik")
                    if div:
                        h = div.find("li", class_="sb-statistik-heim").find("div", class_="sb-statistik-zahl").get_text(strip=True)
                        a = div.find("li", class_="sb-statistik-gast").find("div", class_="sb-statistik-zahl").get_text(strip=True)
                        return h, a
                return "", ""

            # Doelpogingen
            dp_h, dp_a = get_stat(soup_stats, "Doelpogingen")
            doelpogingen_home.append(dp_h)
            doelpogingen_away.append(dp_a)

            # Schoten naast
            sn_h, sn_a = get_stat(soup_stats, "Schoten")
            schoten_naast_home.append(sn_h)
            schoten_naast_away.append(sn_a)
            
            # Hoekschoppen (AANGEPAST: Zoekt nu naar "Corners")
            ck_h, ck_a = get_stat(soup_stats, "Corners")
            hoekschoppen_home.append(ck_h)
            hoekschoppen_away.append(ck_a)

            # Overtredingen
            ov_h, ov_a = get_stat(soup_stats, "Overtredingen")
            overtredingen_home.append(ov_h)
            overtredingen_away.append(ov_a)

            # --- MARKTWAARDE (Vorbericht) ---
            time.sleep(0.5)
            resp_ver = requests.get(vergelijken_url, headers=HEADERS)
            soup_ver = BeautifulSoup(resp_ver.text, "html.parser")
            mw_h, mw_a = "", ""
            for cat in soup_ver.find_all("td", class_="facts-comparison__category"):
                if "Totale MW" in cat.get_text(strip=True):
                    value_row = cat.find_next("tr")
                    tds = value_row.find_all("td")
                    if len(tds) >= 2:
                        mw_h = tds[0].get_text(strip=True)
                        mw_a = tds[1].get_text(strip=True)
                    break
            totale_marktwaarde_home.append(mw_h)
            totale_marktwaarde_away.append(mw_a)

            # --- OPSTELLING ---
            time.sleep(0.5)
            resp_op = requests.get(opstelling_url, headers=HEADERS)
            soup_op = BeautifulSoup(resp_op.text, "html.parser")
            
            mvs_found = []
            for tr in soup_op.find_all("tr"):
                for td in tr.find_all("td"):
                    txt = td.get_text(strip=True)
                    if txt.startswith("Totale MW:"):
                        mvs_found.append(txt.split(":", 1)[1].strip())
            
            while len(mvs_found) < 4:
                mvs_found.append("")

            opstelling_home.append(mvs_found[0])
            opstelling_away.append(mvs_found[1])
            bank_home.append(mvs_found[2])
            bank_away.append(mvs_found[3])

        except Exception as e:
            print(f"Fout bij match detail scraping: {e}")
            # Append lege strings om lijsten synchroon te houden
            doelpogingen_home.append("")
            doelpogingen_away.append("")
            schoten_naast_home.append("")
            schoten_naast_away.append("")
            hoekschoppen_home.append("")
            hoekschoppen_away.append("")
            overtredingen_home.append("")
            overtredingen_away.append("")
            totale_marktwaarde_home.append("")
            totale_marktwaarde_away.append("")
            opstelling_home.append("")
            opstelling_away.append("")
            bank_home.append("")
            bank_away.append("")

    # --- DATA SAMENVOEGEN ---
    for i in range(len(homes)):
        try:
            row = {
                "Speeldag": speeldag,
                "Weekdag": weekdays[i] if i < len(weekdays) else "",
                "Uur": times[i] if i < len(times) else "",
                "Thuisteam": homes[i],
                "Uitteam": aways[i],
                "Thuisdoelpunten": scores_home[i] if i < len(scores_home) else 0,
                "Uitdoelpunten": scores_away[i] if i < len(scores_away) else 0,
                "Doelsaldo": (scores_home[i] - scores_away[i]) if i < len(scores_home) else 0,
                "Toeschouwers": attendees[i] if i < len(attendees) else "",
                "Scheidsrechter": referees[i] if i < len(referees) else "",
                "Rood Thuis": home_reds[i] if i < len(home_reds) else 0,
                "Rood Uit": away_reds[i] if i < len(away_reds) else 0,
                "Positie Thuis": homesRankings[i] if i < len(homesRankings) else "",
                "Positie Uit": awaysRankings[i] if i < len(awaysRankings) else "",
                # Stats
                "Doelpogingen Thuis": doelpogingen_home[i] if i < len(doelpogingen_home) else "",
                "Doelpogingen Uit": doelpogingen_away[i] if i < len(doelpogingen_away) else "",
                "Schoten Naast Thuis": schoten_naast_home[i] if i < len(schoten_naast_home) else "",
                "Schoten Naast Uit": schoten_naast_away[i] if i < len(schoten_naast_away) else "",
                "Hoekschoppen Thuis": hoekschoppen_home[i] if i < len(hoekschoppen_home) else "",
                "Hoekschoppen Uit": hoekschoppen_away[i] if i < len(hoekschoppen_away) else "",
                "Overtredingen Thuis": overtredingen_home[i] if i < len(overtredingen_home) else "",
                "Overtredingen Uit": overtredingen_away[i] if i < len(overtredingen_away) else "",
                # Marktwaarden
                "MW Thuis Totaal": totale_marktwaarde_home[i] if i < len(totale_marktwaarde_home) else "",
                "MW Uit Totaal": totale_marktwaarde_away[i] if i < len(totale_marktwaarde_away) else "",
                "MW Opstelling Thuis": opstelling_home[i] if i < len(opstelling_home) else "",
                "MW Opstelling Uit": opstelling_away[i] if i < len(opstelling_away) else "",
                "MW Bank Thuis": bank_home[i] if i < len(bank_home) else "",
                "MW Bank Uit": bank_away[i] if i < len(bank_away) else "",
            }
            matchday_data.append(row)
        except IndexError:
            continue
            
    return matchday_data

if __name__ == "__main__":
    all_data = []
    
    # LOOP: Speeldag 30 tot en met 30
    print("Start scraping speeldag 30 t/m 31. Dit duurt 20-30 minuten...")
    
    for speeldag in range(30, 31):
        data = scrape_matchday(speeldag)
        all_data.extend(data)
        # Veiligheidspauze tussen speeldagen
        time.sleep(3)

    # Exporteer naar Excel
    if all_data:
        df = pd.DataFrame(all_data)
        output_filename = "transfermarkt_jpl_2025-2026_deel2_speeldag30.xlsx"
        df.to_excel(output_filename, index=False)
        print(f"Klaar! Data opgeslagen in {output_filename}")
    else:
        print("Geen data gevonden of er is een fout opgetreden.")
```
